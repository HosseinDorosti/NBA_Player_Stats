```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nba_data = pd.read_csv("NBA_Player_Stats.csv")
```

```python
# Display the first few rows of the dataset
nba_data.head()

```

```python
# Get summary statistics
nba_data.describe()

```

```python
# Check for missing values
nba_data.isnull().sum()
```

```python
top_scorers = nba_data.sort_values(by='PTS', ascending=False).head(10)
top_scorers
```

```python
position_distribution = nba_data['Pos'].value_counts()
position_distribution
```

```python
#Q: How does age correlate with points scored?
sns.scatterplot(x='Age', y='PTS', data=nba_data)
```

```python
#Q: Which team has the highest average field goal percentage (FG%)?
team_fg_percentage = nba_data.groupby('Tm')['FG%'].mean().sort_values(ascending=False)
team_fg_percentage
```

```python
# Plotting a bar chart for team field goal percentage
plt.figure(figsize=(12, 6))
sns.barplot(x=team_fg_percentage.index, y=team_fg_percentage.values)
plt.title('Average Team Field Goal Percentage')
plt.xlabel('Team')
plt.ylabel('Average FG%')
plt.show()
```

```python
#Q: What is the distribution of player ages in the dataset?
plt.figure(figsize=(10, 6))
sns.histplot(nba_data['Age'], bins=20, kde=True)
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q: How are points (PTS) distributed across different positions?
plt.figure(figsize=(12, 6))
sns.boxplot(x='Pos', y='PTS', data=nba_data)
plt.title('Distribution of Points by Position')
plt.xlabel('Position')
plt.ylabel('Points (PTS)')
plt.show()
```

```python
#Q: What is the relationship between minutes played (MP) and points (PTS)?
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MP', y='PTS', data=nba_data)
plt.title('Relationship between Minutes Played and Points')
plt.xlabel('Minutes Played (MP)')
plt.ylabel('Points (PTS)')
plt.show()
```

```python
#Q: Which teams have the highest average number of blocks (BLK)?
team_avg_blocks = nba_data.groupby('Tm')['BLK'].mean().sort_values(ascending=False)
team_avg_blocks
```

```python
#Q: How does the three-point percentage (3P%) vary across different positions?
plt.figure(figsize=(12, 6))
sns.boxplot(x='Pos', y='3P%', data=nba_data)
plt.title('Three-Point Percentage by Position')
plt.xlabel('Position')
plt.ylabel('Three-Point Percentage (3P%)')
plt.show()
```

```python
#Q: Who are the top players in terms of field goal percentage (FG%) for each position?
top_fg_percentage_by_pos = nba_data.groupby('Pos').apply(lambda x: x.loc[x['FG%'].idxmax()])
top_fg_percentage_by_pos
```

```python
#Q: How many players are there in each position?
position_counts = nba_data['Pos'].value_counts()
position_counts
```

```python
#Q: Is there any correlation between assists (AST) and steals (STL)?
plt.figure(figsize=(10, 6))
sns.scatterplot(x='AST', y='STL', data=nba_data)
plt.title('Correlation between Assists and Steals')
plt.xlabel('Assists (AST)')
plt.ylabel('Steals (STL)')
plt.show()
```
